```python
from tensorboard.backend.event_processing import event_accumulator
import matplotlib.pyplot as plt
def extract_tensorboard_data(logdir):
    # 创建EventAccumulator对象并指定TensorBoard日志目录
    ea = event_accumulator.EventAccumulator(logdir)
    ea.Reload()

    # 获取所有标签（Summary的标签，如Scalar、Image等）
    tags = ea.Tags()
    print(tags)
    # 获取标签为'Scalar'的所有事件
    scalar_events = ea.Scalars('epoch')

    # 从事件中提取数据
    data = {'step': [], 'value': []}
    for event in scalar_events:
        data['step'].append(event.step)
        data['value'].append(event.value)

    return data

# 指定TensorBoard日志目录
tensorboard_logdir = ['version_6','version_7','version_3','version_8']
maxepoch=['learning rate = 1e-3','learning rate = 1e-4','learning rate = 1e-5','learning rate = 1e-6']
tensorboard_logdir
maxepoch
i=0
for path in tensorboard_logdir:
# 提取数据
    tensorboard_data = extract_tensorboard_data(path)

    # 打印数据
    print(tensorboard_data)
    plt.plot(tensorboard_data['step'],tensorboard_data['value'],label=maxepoch[i])
    i+=1
plt.legend()
plt.title('Epochs Over Steps')
plt.xlabel('Step')
plt.ylabel('Epoch')

# 显示网格
plt.grid(True)
```
